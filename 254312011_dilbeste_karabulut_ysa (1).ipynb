```python
# BREAST CANCER WISCONSIN veri seti Ã¼zerinde Ã§alÄ±ÅŸÄ±ldÄ±

# KÃ¼tÃ¼phanelerin yÃ¼klenmesi

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import optuna
import shap
from scipy.stats import zscore
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
"""1. Veri Setinin YÃ¼klenmesi
1.1 scikit-learnâ€™den Veri Seti YÃ¼kleme
KullanÄ±labilecek veri setleri:
â€¢	Breast Cancer Wisconsin (bunu kullandÄ±m)
â€¢	Wine Classification"""

# Veri setini yÃ¼kle
cancer = load_breast_cancer()
```

```python
# Ã¶zellik isimlerini yani sÃ¼tun baÅŸlÄ±klarÄ±nÄ± listeledik
# Ã–zellik isimlerini (feature_names) iÃ§eren listeyi al
feature_names = cancer.feature_names

# Ã–zellik isimlerini ekrana yazdÄ±r
print("## ğŸ§¬ Breast Cancer Wisconsin Veri Seti Ã–zellikleri (Feature Names)")
print("-" * 55)

# NumaralandÄ±rÄ±lmÄ±ÅŸ liste olarak yazdÄ±rma
for i, feature in enumerate(feature_names, 1):
    print(f"{i:02}. {feature}")

print("-" * 55)
print(f"\nToplam Ã–zellik SayÄ±sÄ±: {len(feature_names)}")
```

```python
# hedef isimlerini yani sÃ¼tun baÅŸlÄ±klarÄ±nÄ± listeledik

# Ã–zellik isimlerini (feature_names) iÃ§eren listeyi al
target_names = cancer.target_names
#target_names2 = cancer.target
#print("target:",target_names2)
# Ã–zellik isimlerini ekrana yazdÄ±r
print("## ğŸ§¬ Breast Cancer Wisconsin Veri Seti Hedef Ä°simleri (Target Names)")
print("-" * 55)

# NumaralandÄ±rÄ±lmÄ±ÅŸ liste olarak yazdÄ±rma
for i, target in enumerate(target_names, 1):
    print(f"{i:02}. {target}")

print("-" * 55)
print(f"\nToplam Hedef SayÄ±sÄ±: {len(target_names)}")
```
